Загрузите в колаб файлы по оценкам (ratings) и фильмам (movies) и создайте на их основе pandas-датафреймы

```python
import pandas as pd
```

```python
df_ratings=pd.read_csv("u.data.csv",sep='\t',names=["user id","item id","rating","timestamp"])
```

```python
df_movies=pd.read_csv("u.item.csv",sep='|',encoding='ISO-8859-1',names=["movie id","movie title","release date","video release date","IMDb URL","unknown","Action"," Adventure","Animation","Children's","Comedy","Crime","Documentary","Drama","Fantasy","Film-Noir","Horror","Musical","Mystery","Romance","Sci-Fi","Thriller","War","Western"])
```

Средствами Pandas, используя dataframe ratings, найдите id пользователя, поставившего больше всего оценок

```python
df_ratings['user id'].value_counts()
```

Оставьте в датафрейме ratings только те фильмы, который оценил данный пользователь 

```python
df_r = df_ratings[ df_ratings['user id']==405  ]
df_r.head()
```

**Добавьте к датафрейму из задания 3 столбцы:**

●	**По жанрам. Каждый столбец - это жанр. Единицу записываем, если фильм ** **принадлежит данному жанру и 0 - если нет **

●	**столбцы с общим количеством оценок от всех пользователей на фильм и суммарной**
 **оценкой от всех пользователей **

```python
df_r.rename(columns={'item id': 'movie id'}, inplace=True)
df_r.head(5)
```

```python
df_r.info()
```

```python
df_m = df_r.merge(df_movies, how='left', on='movie id')
df_m.head()
```

**● столбцы с общим количеством оценок от всех пользователей на фильм и суммарной оценкой от всех пользователей**

```python
df_count=df_ratings.groupby('user id').count().reset_index()
df_count=df_count[['user id','rating']]
df_count.rename(columns={'rating': 'count','user id':'movie id'}, inplace=True)
df_count.head(4)
```

```python
df_sum=df_ratings.groupby('user id').sum().reset_index()
df_sum=df_sum[['user id','rating']]
df_sum.rename(columns={'rating': 'sum','user id':'movie id'}, inplace=True)
df_sum.head(4)
```

```python
df_resul = df_m.merge(df_sum,left_on="movie id",right_on="movie id",how="inner")
df_resul.head(3)
```

```python
df_resul = df_resul.merge(df_count,left_on="movie id",right_on="movie id",how="inner")
df_resul.head(3)
```

```python
df_resul['release date']=pd.to_datetime(df_resul['release date'],errors='coerce')
```

```python
df_resul['year']=df_resul['release date'].dt.year
```

```python
df_resul['month']=df_resul['release date'].dt.month
df_resul.head(3)
```

Сформируйте X_train, X_test, y_train, y_test

```python
df_resul.columns
```

```python
X, y = df_resul[['user id', 'movie id',  'timestamp',  'unknown', 'Action',
       ' Adventure', 'Animation', "Children's", 'Comedy', 'Crime',
       'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical',
       'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western', 'sum',
       'count', 'year', 'month']], df_resul['rating']
```

```python
X
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train.head()
```

**Возьмите модель линейной регрессии (или любую другую для задачи регрессии)  и обучите ее на фильмах**

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

**Оцените качество модели на X_test, y_test при помощи метрик для задачи регрессии**

```python
mean_squared_error (y_test, model.predict(X_test))
```

**Посчитайте средствами спарка среднюю оценку для каждого жанра**

```python
test = PYratings.withColumn('unknown', when(col('unknown').isin('1'), "unknown")\
                            .otherwise(when(col('unknown').isin('0'), "")
                            .otherwise(PYratings.unknown)))
test = test.withColumn('Action', when(col('Action').isin('1'), "Action")\
                            .otherwise(when(col('Action').isin('0'), "")
                            .otherwise(test.Action)))
test = test.withColumn('Adventure', when(col('Adventure').isin('1'), "Adventure")\
                            .otherwise(when(col('Adventure').isin('0'), "")
                            .otherwise(test.Adventure)))
test = test.withColumn('Animation', when(col('Animation').isin('1'), "Animation")\
                            .otherwise(when(col('Animation').isin('0'), "")
                            .otherwise(test.Animation)))
test = test.withColumn('Children', when(col('Children').isin('1'), "Children")\
                            .otherwise(when(col('Children').isin('0'), "")
                            .otherwise(test.Children)))
test = test.withColumn('Comedy', when(col('Comedy').isin('1'), "Comedy")\
                            .otherwise(when(col('Comedy').isin('0'), "")
                            .otherwise(test.Comedy)))
test = test.withColumn('Crime', when(col('Crime').isin('1'), "Crime")\
                            .otherwise(when(col('Crime').isin('0'), "")
                            .otherwise(test.Crime)))
test = test.withColumn('Documentary', when(col('Documentary').isin('1'), "Documentary")\
                            .otherwise(when(col('Documentary').isin('0'), "")
                            .otherwise(test.Documentary)))
test = test.withColumn('Drama', when(col('Drama').isin('1'), "Drama")\
                            .otherwise(when(col('Drama').isin('0'), "")
                            .otherwise(test.Drama)))
test = test.withColumn('Fantasy', when(col('Fantasy').isin('1'), "Fantasy")\
                            .otherwise(when(col('Fantasy').isin('0'), "")
                            .otherwise(test.Fantasy)))
test = test.withColumn('Film_Noir', when(col('Film_Noir').isin('1'), "Film_Noir")\
                            .otherwise(when(col('Film_Noir').isin('0'), "")
                            .otherwise(test.Film_Noir)))
test = test.withColumn('Horror', when(col('Horror').isin('1'), "Horror")\
                            .otherwise(when(col('Horror').isin('0'), "")
                            .otherwise(test.Horror)))
test = test.withColumn('Musical', when(col('Musical').isin('1'), "Musical")\
                            .otherwise(when(col('Musical').isin('0'), "")
                            .otherwise(test.Musical)))
test = test.withColumn('Mystery', when(col('Mystery').isin('1'), "Mystery")\
                            .otherwise(when(col('Mystery').isin('0'), "")
                            .otherwise(test.Mystery)))
test = test.withColumn('Romance', when(col('Romance').isin('1'), "Romance")\
                            .otherwise(when(col('Romance').isin('0'), "")
                            .otherwise(test.Romance)))
test = test.withColumn('Sci_Fi', when(col('Sci_Fi').isin('1'), "Sci_Fi")\
                            .otherwise(when(col('Sci_Fi').isin('0'), "")
                            .otherwise(test.Sci_Fi)))
test = test.withColumn('Thriller', when(col('Thriller').isin('1'), "Thriller")\
                            .otherwise(when(col('Thriller').isin('0'), "")
                            .otherwise(test.Thriller)))
test = test.withColumn('War', when(col('War').isin('1'), "War")\
                            .otherwise(when(col('War').isin('0'), "")
                            .otherwise(test.War)))
test = test.withColumn('Western', when(col('Western').isin('1'), "Western")\
                            .otherwise(when(col('Western').isin('0'), "")
                            .otherwise(test.Western)))
```

```python
test.show()
```

```python
test1 = test.select(concat_ws('', test.unknown, test.Action, test.Adventure, test.Animation, test.Children, test.Comedy, test.Crime,\
                              test.Documentary, test.Drama, test.Fantasy, test.Film_Noir, test.Horror, test.Musical, test.Mystery,\
                              test.Romance, test.Sci_Fi, test.Thriller, test.War, test.Western).alias('Genre'), test.Rating)
```
